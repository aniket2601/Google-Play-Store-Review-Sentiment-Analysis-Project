```python
new.drop([10472],inplace = True)
```

```python
# Defining the function to fill the null values with the median.
def impute_median(series):
  return series.fillna(series.median())
```

```python
new.Rating = new['Rating'].transform(impute_median)
```

```python
new.isnull().sum()
```

```python
# Mode of categorical values
print(new['Type'].mode())
print(new['Current_Ver'].mode())
print(new['Android_Ver'].mode())
```

```python
# Filling the missing categorical values with Mode
new['Type'].fillna(str(new['Type'].mode().values[0]),inplace=True)
new['Current_Ver'].fillna(str(new['Current_Ver'].mode().values[0]),inplace=True)
new['Android_Ver'].fillna(str(new['Android_Ver'].mode().values[0]),inplace=True)
```

```python
# Counting the number of null values in each column
new.isnull().sum()
```

```python
# Converting Price,Reviews,Ratings into numerical values
new['Price'] = new['Price'].apply(lambda x : str(x).replace('$','') if '$' in str(x) else str(x))
new['Price'] = new['Price'].apply(lambda x : float(x))
new['Reviews'] = pd.to_numeric(new['Reviews'],errors='coerce')
```

```python

new['Installs'] = new['Installs'].apply(lambda x : str(x).replace('+','') if '+' in str(x) else str(x))
new['Installs'] = new['Installs'].apply(lambda x : str(x).replace(',','') if ',' in str(x) else str(x))
new['Installs'] = new['Installs'].apply(lambda x : float(x))
```

```python

# Creating category for categorical data
grp = new.groupby('Category')
x = grp['Rating'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
def outliers(new):
    a = new.select_dtypes(include = [np.number]).columns.tolist()
    for i in a:
        q1 = new[i].quantile(0.25)
        q3 = new[i].quantile(0.75)
        iqr = q3 - q1
        upper_limit = q3+1.5*iqr
        lower_limit = q1-1.5*iqr
        new[i] = np.where(new[i]>upper_limit,upper_limit,new[i])
        new[i] = np.where(new[i]<lower_limit,lower_limit,new[i])
```

```python
outliers(new)
```
